## Gradient descent
* Let $\boldsymbol{Y}\in\mathbb{R}^{d\times d}$ be a blurred version
of the image $\boldsymbol{X}\in\mathbb{R}^{d\times d}$:
$$\boldsymbol{Y}=\boldsymbol{H}\boldsymbol{X}\boldsymbol{H}^{T}$$
where $\boldsymbol{H}\in\mathbb{R}^{d\times d}$ is a known (separable)
blurring matrix.

#### Load the data ($\boldsymbol{Y}$ and $\boldsymbol{H}$):

````python
from scipy.io import loadmat

data = loadmat('data.mat')
mY   = data['Y']
mH   = data['H']

mY.shape, mH.shape
````

#### Plot $Y$ and $H$:

````python
_, ax = plt.subplots(1, 2, figsize=(12, 6))
ax[0].imshow   (mY, cmap='gray')
ax[1].matshow  (mH)
ax[0].set_title('$Y = H X H^T$')
ax[1].set_title('$H$')
plt.show()
````

* Our goal is to find $\boldsymbol{X}$ that minimizes:
$$\boldsymbol{X}^{\star}=\arg\min_{\boldsymbol{X}}f\left(\boldsymbol{X}\right)=\arg\min_{\boldsymbol{X}}\left\Vert \boldsymbol{Y}-\boldsymbol{H}\boldsymbol{X}\boldsymbol{H}^{T}\right\Vert _{F}^{2}$$

### 2.1
1. Find a closed form expression for $\boldsymbol{X}^{\star}$ as a function of $\boldsymbol{Y}$ and $\boldsymbol{H}$.
2. Compute $\boldsymbol{X}^{\star}$ and plot it.

### 2.1   -   Solution

````python
#-- This solution without optimization and we can use inverse matrix

Xstar = np.linalg.inv(mH)@mY@np.linalg.inv(mH.T)                   # The solution x=(H^-1)*Y*(H.T^-1)* 

_, ax = plt.subplots(1, 2, figsize=(12, 6))                        # Plot the results
ax[0].imshow   (mY,     cmap='gray')
ax[1].imshow   (Xstar, cmap='gray')
ax[0].set_title('$Y = H X H^T$')
ax[1].set_title('$X^{\star}$')
plt.show()
````

### 2.2   -   Solution

````python
import time                                                    # time calculation library

#==================================================#
def f(X):
    '''
    f(X) = ||Y - HXH^T||_F^2 is the objective function
    '''
    f = np.linalg.norm(mY-mH@X@mH.T, ord='fro')**2
    return f
#==================================================#
#==================================================#
def Df(X):
    '''
    Df(X) = f(X) is the objective gradient
    '''
    Df = -2*mH.T@(mY-mH@X@mH.T)@mH
    return Df
#==================================================#
#==================================================#
def ApplyGradientDescent(f, Df, X0, lr, ):
    '''
    Apply gradient descent.
    Args:
        f  - Objective function
        Df - Objective gradient
        X0 - Initial point
        lr - Learning rate (step size)
          - Stopping criterion value
    Output:
        X  - The converged X
    Stopping criterion:
        Stop iterate when f(X) < 
    '''
    i=1                                            # Start with counter i=1  
    X=X0                                           # Initial point
    t1 = time.time()                               # Initial time

    while (f(X)>=):
        print ("The Iteration number is", i, ",  The f(x) value is", f(X))
        X  = X - (lr)*Df(X)                        # Gradient decent calculation
        i=i+1                                      # Count number of iteration
        
    print("\n  Total time is",time.time() - t1)    # Print total time at the end
    return X
#==================================================#
````

````python
#-- Plot the results

"""
few tests before:
for lr=0.3, =1e-2 we get 1154 iterations, for lr=0.5, =1e-2 we get 692 iterations, 
for lr=0.95, =1e-2 we get 364 iterations, for lr=1, =1e-2 we get 360 iterations, for lr>1 f(X)=inf
for lr=1, =1e-3 we get 2333 iterations, for lr=1, =1e-4 we get 17613 iterations
"""
# Final  -  for lr=1, =1e-4 we get 17613 iterations, total time is 155.8 seconds

X = ApplyGradientDescent(f, Df, X0=mY, lr=1, =1e-4)       # Call main function with initial values

_, ax = plt.subplots(1, 2, figsize=(12, 6))
ax[0].imshow   (mY,     cmap='gray')
ax[1].imshow   (X , cmap='gray')
ax[0].set_title('$Y = H X H^T$')
ax[1].set_title('$final obtained X$')
plt.show()
````

### Momentum
<div class="alert alert-block alert-warning">
<span style="color:black">
    
* <ins>Momentum</ins> is an effective method to improve convergence rate.
* Gradient descent update rule:
$$\boldsymbol{x}^{\left(t\right)}=\boldsymbol{x}^{\left(t-1\right)}-\mu\nabla f\left(\boldsymbol{x}^{\left(t-1\right)}\right)$$
* Momentum update rule:
$$\boldsymbol{x}^{\left(t\right)}=\boldsymbol{x}^{\left(t-1\right)}-\mu\nabla f\left(\boldsymbol{x}^{\left(t-1\right)}\right)+\beta\left(\boldsymbol{x}^{\left(t-1\right)}-\boldsymbol{x}^{\left(t-2\right)}\right)$$
where $\beta\in[0,1)$ and typically $0.9\leq\beta\leq0.99$.

</span>
</div>

### 2.3   -   Solution

````python
import time                                                    # time calculation library

#==================================================#
def f(X):
    '''
    f(X) = ||Y - HXH^T||_F^2 is the objective function
    '''
    f = np.linalg.norm(mY-mH@X@mH.T, ord='fro')**2
    return f
#==================================================#
#==================================================#
def Df(X):
    '''
    Df(X) = f(X) is the objective gradient
    '''
    Df = -2*mH.T@(mY-mH@X@mH.T)@mH
    return Df
#==================================================#
#==================================================#

def ApplyGradientDescentMomentum(f, Df, X0, lr, , ):
    '''
    Apply gradient descent with momentum.
    Args:
        f  - Objective function
        Df - Objective gradient
        X0 - Initial point
        lr - Learning rate (step size)
          - Momentum decaying factor, [0,1)  and typically 0.90.99.
          - Stopping criterion value
    Output:
        X  - The converged X
    Stopping criterion:
        Stop iterate when f(X) < 
    '''
    i  = 1                                         # Start with counter i=1  
    X  = X1 = X0                                   # Initial points
    X1 = X1 - (lr)*Df(X1) + *(X1-X0)              # Initial points 
    
    t1 = time.time()                               # Initial time

    while (f(X)>=):
        print ("The Iteration number is", i, ",  The f(x) value is", f(X))
        X  = X1 -  (lr)*Df(X1)  + *(X1-X0)        # Gradient decent with momentum calculation
        X0 = X1                                    # promote the i-2 point 
        X1 = X                                     # promote the i-1 point  
        i=i+1                                      # Count number of iteration
        
    print("\n  Total time is",time.time() - t1)    # Print total time at the end
    return X
````

````python
#-- Plot the results

"""
few tests before:
for lr=1, =1e-4, =0.9 we get 1760 iterations, total time is 15.3 seconds,
for lr=1, =1e-4, =0.94 we get 1055 iterations, total time is 9.2 seconds,
for lr=1, =1e-4, =0.95 we get 879 iterations, total time is 7.8 seconds,
for lr=1, =1e-4, =0.96 we get 703 iterations, total time is 6.12 seconds,
for lr=1, =1e-4, =0.97 we get 535 iterations, total time is 4.9 seconds,
for lr=1, =1e-4, =0.98 we get 642 iterations, total time is 5.5 seconds,
for lr=1, =1e-4, =0.99 we get 1232 iterations, total time is 10.9 seconds,

Further we paid attention that lr can be more than 1 (< 2)
"""
# Final  -  for lr=1.9, =1e-4, =0.96 we get 388 iterations, total time is 3.5 seconds

X = ApplyGradientDescentMomentum(f, Df, X0=mY, lr=1.9, =0.96, =1e-4)       # Call main function with initial values

_, ax = plt.subplots(1, 2, figsize=(12, 6))
ax[0].imshow   (mY,     cmap='gray')
ax[1].imshow   (X , cmap='gray')
ax[0].set_title('$Y = H X H^T$')
ax[1].set_title('final obtained $X$')
plt.show()
````
